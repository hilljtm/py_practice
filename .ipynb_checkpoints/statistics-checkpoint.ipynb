```python
# Plot all your data

# A major drawback of histograms is that the same dataset can look different depending on the number of bins
# This leads to binning bias
# Binning bias: same data may be interpreted differently depending on the choice of bins
# Histograms don't plot ALL data, your sweeping data into bins and losing the actual values
# To remedy this we can use a swarm plot
# In order to plot a swarm plot the data needs to be in a data frame
# each column is a feature of interest, each row is an observation
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Import your data
df = pd.read_csv('swing_states.csv')
df.head()
```

```python
# Binning bias
plt.hist(df['dem_share'])
plt.xlabel('percentage of votes for obama')
plt.ylabel('number of counties')
plt.show();

sns.set()
plt.hist(df['dem_share'], bins=20)
plt.xlabel('percentage of votes for obama')
plt.ylabel('number of counties')
plt.show();

sns.set()
plt.hist(df['dem_share'], bins=25)
plt.xlabel('percentage of votes for obama')
plt.ylabel('number of counties')
plt.show();
```

```python
sns.swarmplot(x='state', y='dem_share', data=df)
plt.xlabel('state')
plt.ylabel('percentage of votes for obama')
plt.show();
```

```python
# Empirical cumulative distribution function (ECDF)
# An X-value of an ECDF is the quantity you are measuring
# The Y-value is the function of data points that have a value smaller than the corresponding x-value

x = np.sort(df['dem_share'])
y = np.arange(1, len(x) + 1) / len(x)
plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('percentage of votes for obama')
plt.ylabel('ECDF')
plt.margins(0.02)
plt.show();
```

```python
# Computing the ECDF by writing your own function

def ecdf(df):
    """Compute ECDF for a one-dimensional array of measurements"""
    # Number of data points: n
    n = len(df)
    # X-data for ECDF: x
    x = np.sort(df)
    # Y-data for ECDF: y
    y = np.arange(1, n + 1) / n
    return x, y
```

```python
# Plot the ECDF using your function
# You will need to unpack your arrays, since the function returns 2
df = pd.read_csv('iris_data.csv')


x_vers, y_vers = ecdf(df['petal length(cm)'])
plt.plot(x_vers, y_vers, marker='.', linestyle='none')
plt.xlabel('versicolor petal length')
plt.ylabel('ECDF')
plt.show()
```
